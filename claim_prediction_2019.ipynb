```python
#importing the training and testing dataset 
def alldata():
    #read the data
    train = pd.read_csv('train_data.csv')
    test = pd.read_csv('test_data.csv')
    
    #Extracting and then removing the targets from the training data
    response = train.Claim
    train.drop(['Claim'], 1, inplace=True)
    
    #Combining the train data and test data for feature engineering
    dataset = train.append(test)
    dataset.reset_index(inplace=True)
    
    #So i will drop the Customer Id 
    dataset.drop(['index', 'Customer Id'], inplace = True, axis=1)
    
    return dataset
```

```python
dataset = alldata()
```

```python
def clip(x):
    if x <= 0.5:
        return 0.5
    else:
        return 1
dataset['Insured_Period'] = dataset['Insured_Period'].apply(clip)
```

```python
# Assigning N-painted = 1 & V-notPaintd = 0
dataset['Building_Painted'] = np.where(dataset['Building_Painted'] == 'N', 1, dataset['Building_Painted'])
dataset['Building_Painted'] = np.where(dataset['Building_Painted'] == 'V', 0, dataset['Building_Painted'])
dataset['Building_Painted'] = dataset.Building_Painted.astype('int64')
```

```python
# Assigning N-Fenced = 1 & V-Not_Fenced = 0
dataset['Building_Fenced'] = np.where(dataset['Building_Fenced'] == 'N', 1, dataset['Building_Fenced'])
dataset['Building_Fenced'] = np.where(dataset['Building_Fenced'] == 'V', 0, dataset['Building_Fenced'])
dataset['Building_Fenced'] = dataset.Building_Fenced.astype('int64')
```

```python
# Assigning V-Has_Garden = 1 & O-No_Garden = 0
dataset['Garden'] = np.where(dataset['Garden'] == 'V', 1, dataset['Garden'])
dataset['Garden'] = np.where(dataset['Garden'] == 'O', 0, dataset['Garden'])
```

```python
from sklearn.model_selection import train_test_split
#Train-Test-Split
X_train, X_test, y_train, y_test = train_test_split(train,targets, test_size=0.35, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
def makeOverSamplesSMOTE(X,y):
 #input DataFrame
 #X →Independent Variable in DataFrame\
 #y →dependent Variable in Pandas DataFrame format
 from imblearn.over_sampling import SMOTE
 sm = SMOTE()
 X, y = sm.fit_sample(X, y)
 return X,y
```

```python
makeOverSamplesSMOTE(train,targets)
```

```python
from xgboost import XGBClassifier
```

```python
clf = XGBClassifier(base_score=0.7,n_estimator=1000,booster='dart',max_depth=8,objective='binary:logistic',learning_rate=0.01,subsample=0.9,reg_lambda=0.03)
clf.fit(X_train, y_train, eval_set=None, eval_metric="auc")
```

```python
#Checking Training performance on the XGBoost Model
y_train_pred = clf.predict(X_train)
count_misclass = (y_train != y_train_pred).sum()
print('No. of Misclassified samples: {}'.format(count_misclass))
auc_roc = roc_auc_score(y_train_pred, y_train)
print('auc_roc: {:.5f}'.format(auc_roc))
```

```python
#Checking Testing performance on the XGBoost Model
y_test_pred = clf.predict(X_test)
count_misclass = (y_test != y_test_pred).sum()
print('No. of Misclassified samples: {}'.format(count_misclass))
auc_roc = roc_auc_score(y_test_pred, y_test)
print('auc_roc: {:.5f}'.format(auc_roc))
```
